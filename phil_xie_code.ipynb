```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('chestmnist.npz')
```

```python
data.files
```

# 1. Grayscale image

```python
img = data['train_images'][1]
```

```python
plt.imshow(img, cmap='gray') # bone colormap just for fun
plt.savefig("phil_xie_1.png")
```

```python
plt.imshow(img, cmap='bone') # bone colormap just for fun
```

# 2. Invert

```python
img_inverted = np.array([255 - x for x in img])
```

```python
plt.imshow(img_inverted, cmap='gray')
plt.savefig("phil_xie_2.png")
```

# 3. Other filter

```python
from scipy import ndimage
edge_hori = ndimage.sobel(img, 0)
edge_vert = ndimage.sobel(img, 1)
```

```python
plt.imshow(edge_hori, cmap="gray")
plt.savefig("phil_xie_3a.png")
```

```python
plt.imshow(edge_vert, cmap="gray")
plt.savefig("phil_xie_3b.png")
```
